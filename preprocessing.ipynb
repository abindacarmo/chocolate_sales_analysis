```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

data = pd.read_csv("chocolate_dataset/Chocolate-Sales.csv")

chocolate_data = data.head(100)

chocolate_data.info()
chocolate_data.describe()
chocolate_data.isnull().sum()
choco_clean = (chocolate_data.dropna().drop_duplicates())
```

<h3>Vizualization</h3>

```python
# Pie chart Product
product_counts = choco_clean['Product'].value_counts()
plt.figure(figsize=(30,20))
plt.pie(product_counts, labels=product_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Produk sales')
plt.show()
```

```python
# Pie chart Boxes Shipped
product_counts = choco_clean['Boxes Shipped'].value_counts()
plt.figure(figsize=(30,20))
plt.pie(product_counts, labels=product_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Boxes Shipped Sales')
plt.show()
```
